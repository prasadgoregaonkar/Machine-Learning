```python
#indexing and Row Selection
#Print the name of the oldest passenger travelling on ship.
#step 1. find max value for age
df['Age'].max()
```

```python
# step 2.extract those age columns value=maxValue
df[df['Age']==df['Age'].max()]['Name']
```

```python
#Sorting
#1)sort_values()
#2)Arguments are-a. By,b. ascending
#sort.values(by=[])
df.sort_values(by='Age',ascending='True')
```

```python
#sort more then one column
df.sort_values(by=['Name','Age'],ascending=[True,False])
```

```python
#Replacing values in column
#1. through map() method
#create dictonary with  old values that to replace
new={'male':0,'female':1}
df['Sex']=df['Sex'].map(new)
df.head()
```

```python
#2. through replace method
#create dictonary with  old values that to replace
old={0:'male',1:'female'}
df.replace({'Sex':old})
```

```python
#Groupby
df.groupby(by='Survived')['Age'].describe()
```

```python
#Contingency Table
pd.crosstab(df['Sex'],df['Survived'],normalize=True)
```

```python
df.pivot_table(['Age','Fare'],['Survived'],aggfunc='mean')
```
